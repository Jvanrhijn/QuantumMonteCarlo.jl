```julia
using Distributions
using Plots
using DataStructures
using Random
using LinearAlgebra
using HDF5
using StatsBase

using QuantumMonteCarlo

reference_energy = -1.14
exact_energy = -1.1714

#include("forceutil.jl")

# DMC settings
τ = 1e-3
nwalkers = 10
num_blocks = 1000
steps_per_block = Int64(1/τ)
neq = 10
a = [-0.7, 0.0, 0.0]
b = [0.7, 0.0, 0.0]


# Trial wave function
function ψslater(x::Array{Float64})
    α = 1.0
    exp(-α*norm(x))
end

function jastrow(x::Array{Float64})
    β = 1.0
    γ = 1.0
    
    x1 = x[1:3]
    x2 = x[4:end]
    r12 = norm(x1 - x2)
    
    exp(β*r12 / (1 + γ*r12))
end

function ψ(x)
    x1 = x[1:3]
    x2 = x[4:end]
    (ψslater(x1 - a) + ψslater(x1 - b)) * (ψslater(x2 - a) + ψslater(x2 - b)) * jastrow(x)
end

ψtrial = WaveFunction(
    ψ,
    x -> gradient_fd(ψ, x),
    x -> laplacian_fd(ψ, x)
)

# Setting up the hamiltonian
function hamiltonian(ψstatus, x)
    x1 = x[1:3]
    x2 = x[4:end]
    ψval = ψstatus.value
    kinetic = -0.5 * ψstatus.laplacian
    potential_en = -1/norm(x1 - a) - 1/norm(x1 - b) - 1/norm(x2 - a) - 1/norm(x2 - b)
    potential_ee = 1/norm(x1 - x2)
    potential_nn = 1/norm(a - b)
    return kinetic + (potential_en + potential_ee + potential_nn) * ψval
end

function hamiltonian_recompute(ψfunc, x)
    x1 = x[1:3]
    x2 = x[4:end]
    ψval = ψfunc.value(x)
    kinetic = -0.5 * ψfunc.laplacian(x)
    potential_en = -1/norm(x1 - a) - 1/norm(x1 - b) - 1/norm(x2 - a) - 1/norm(x2 - b)
    potential_ee = 1/norm(x1 - x2)
    potential_nn = 1/norm(a - b)
    return kinetic + (potential_en + potential_ee + potential_nn) * ψval
end

model = Model(
    hamiltonian,
    hamiltonian_recompute,
    ψtrial,
)

rng = MersenneTwister(0)

# create "Fat" walkers
walkers = QuantumMonteCarlo.generate_walkers(nwalkers, ψtrial, rng, Normal(0, 3), 6)
fat_walkers = [QuantumMonteCarlo.FatWalker(walker, OrderedDict()) for walker in walkers]

### Actually run DMC
energies, errors = QuantumMonteCarlo.run_dmc!(
    model, 
    fat_walkers, 
    τ, 
    num_blocks, 
    steps_per_block, 
    reference_energy; 
    rng=rng, 
    neq=neq, 
    brancher=stochastic_reconfiguration_pyqmc!,
    verbosity=:progressbar,
);
```

```julia
println("Energy: $(last(energies)) +- $(last(errors))")
```

```julia
plot(energies, ribbon=(errors, errors), fillalpha=0.2, label="DMC energy")
hline!([exact_energy], color="grey", labels="Exact energy")
hline!([-1.14], color="black", labels="Reference energy")
```
